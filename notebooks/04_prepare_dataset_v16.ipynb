# 04 - Prepare Physics-Aware Dataset (v16)

## Objective
Prepare the physics-aware urban heat dataset for machine learning by:
- Loading the processed dataset,
- Removing non-predictive columns,
- Checking data quality,
- Creating train-test splits,
- Exporting ML-ready datasets.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(
    "../data/processed/Delhi_Urban_Heat_Dataset_v16.csv"
)

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df = df.drop(
    columns=[
        "system:index",
        ".geo"
    ]
)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df[
    [
        "mean_ndvi",
        "mean_ndbi",
        "mean_ndwi",
        "mean_emissivity",
        "mean_albedo"
    ]
]

y = df["mean_lst"]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
X_train.to_csv(
    "../data/processed/X_train_v16.csv",
    index=False
)

X_test.to_csv(
    "../data/processed/X_test_v16.csv",
    index=False
)

y_train.to_csv(
    "../data/processed/y_train_v16.csv",
    index=False
)

y_test.to_csv(
    "../data/processed/y_test_v16.csv",
    index=False
)
```

```python
print("Dataset preparation complete.")

print("\nFeatures:")
print(X.columns.tolist())

print("\nTraining samples:", len(X_train))
print("Testing samples :", len(X_test))
```
